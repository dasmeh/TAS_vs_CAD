## Part 1: Loading results from FUMA

```r
library("stringr")
library("ggplot2")
```

```r
TAS=read.table("FUMA/TAS_FUMA_GS.txt", sep="\t", header = TRUE)
```

```r
CAD=read.table("FUMA/CAD_FUMA_GS.txt", sep="\t", header = TRUE)
```

```r
TAS_subset=subset(TAS, Category=="GO_bp")
CAD_subset=subset(CAD, Category=="GO_bp")
```

```r
TAS_subset
```

```r
TAS_stat=-log10(TAS_subset$adjP)
CAD_stat=-log10(CAD_subset$adjP[match(TAS_subset$GeneSet, CAD_subset$GeneSet)])
```

```r
plot(TAS_stat, CAD_stat)
```

```r
diff_path_stat=TAS_stat - CAD_stat
names(diff_path_stat)=TAS_subset$GeneSet
```

```r
diff_path_stat_sorted=sort(diff_path_stat, decreasing = TRUE)
```

## Part 2: Ploting the final results

```r
library("stringr")
```

```r
Final_list=read.csv("TAS_vs_CAS_final_selection.csv")
```

```r
Final_list=Final_list[which(nchar(Final_list$Remark)!=0),]
```

```r
MAT=cbind(Final_list$TAS_stat, Final_list$Remark)
rownames(MAT)=Final_list$GO_terms
colnames(MAT)=c("Stat", "Remark")
MAT=as.data.frame(MAT)
MAT[,1]=as.numeric(MAT[,1])
rownames(MAT)=Final_list$GO_terms
```

```r
rownames(MAT)=str_split(rownames(MAT), pattern = "GOBP_", simplify = TRUE)[,2]
```

```r
MAT=subset(MAT, Stat>3)
```

```r
library(ggplot2)

# Assuming your matrix is named my_matri

# Convert matrix to data frame
my_df <- as.data.frame(MAT)

# Plot
ggplot(my_df, aes(x = rownames(MAT), y = Stat, fill = Remark)) +
  geom_bar(stat = "identity") +
  theme_minimal() +
  labs(x = "GO Terms", y = "Statistics", title = "Vertical Bar Plot with Different Colors for Remarks") +
  theme(axis.text.x = element_text(angle = 90, hjust = 1))
```

```r
# Reorder rows based on the descending order of Statistics
my_df <- my_df[order(-my_df$Stat), ]

p <- ggplot(my_df, aes(x = reorder(rownames(MAT), -Stat), y = Stat, fill = Remark)) +
  geom_bar(stat = "identity") +
  theme_minimal() +
  labs(x = "GO Terms", y = "Statistics", title = "Vertical Bar Plot with Different Colors for Remarks") +
  theme(axis.text.x = element_text(angle = 20, hjust = 1.2))
p

```
